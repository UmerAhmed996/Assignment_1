```python
def check_pallin(number):
    t = number
    rev =0
    
    while(number > 0):
        d = number % 10
        rev = rev * 10 + d
        number = number // 10
    if(t == rev):
        return True
    else:
        return False

Totalnumbers = 0

for i in range(10000):
    num = i
    
    for k in range(50):
        num = (int(str(num)[::-1]) + num)
        if check_pallin(num) == 1:
            break
        if k == 49:
            Totalnumbers+=1

print(Totalnumbers)
```
